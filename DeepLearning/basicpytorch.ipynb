```python
np.random.seed(0)
player_1 = np.array([np.random.randint(-10,10),np.random.randint(-10,10)])
```

```python
player_1
```

```python
player_2 = np.array([np.random.randint(-10,10),np.random.randint(-10,10)])
```

```python
player_2
```

```python
res = player_1 + player_2
```

```python
res
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,6))
plt.quiver(0,0,player_1[0],player_1[1], angles='xy', scale_units='xy', scale=1, color='r', label="Player A's Movement")
plt.quiver(0,0,player_2[0],player_2[0],angles = 'xy',scale_units = 'xy',label = "Player_2 Movement",color = 'g')
plt.quiver(0,0,res[0],res[1],angles='xy',scale_units = 'xy',scale = 1,color = 'b',label = 'Res Movement')

plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.axvline(x=0, color='grey', lw=1)
plt.axhline(y=0, color='grey', lw=1)
plt.title('Vector Addition of Players Movements')
plt.xlabel('Horizontal Movement')
plt.ylabel('Vertical Movement')
plt.grid(True)
plt.legend()
plt.show()


```
