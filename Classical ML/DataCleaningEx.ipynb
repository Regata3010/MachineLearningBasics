```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
df
```

```python
df.drop_duplicates() #delete duplbicates
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

```python
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Last_Name"] = df["Last_Name"].fillna("No_Surname")

df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("[^0-9]", "", regex=True)
df

```

```python
df["Phone_Number"]  = df["Phone_Number"].apply(lambda x : str(x))
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x : x[0:3] + "-" + x[3:6] + "-" + x[6:10])
df
```

```python

df["Phone_Number"]  = df["Phone_Number"].str.replace("--","")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan","")
df
```

```python
df["Address"] = df["Address"].str.replace("N/a","")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("N/a","")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].replace({'Y': 'Yes', 'N': 'No'})
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna("")
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({'Y':'Yes','N':'No'})
df
```

```python
df.drop_duplicates(inplace = True)
```

```python
df
```

```python
df.dropna(subset=["Phone_Number","Address"],inplace = True)
df
```

```python
# Drop row with index 2
df.drop(index=18, inplace=True)
```
