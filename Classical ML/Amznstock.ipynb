```python
plt.plot(df['Date'],df['Adj Close'],label='Adj Close')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=6))
plt.gcf().autofmt_xdate()
plt.xticks(rotation=45)
plt.title('Adj Close Price')
plt.legend()
plt.show()
```

```python
df['50_MA'] = df['Close'].rolling(window=50).mean()
df['200_MA'] = df['Close'].rolling(window=200).mean()

plt.plot(df['Close'], label='Close')
plt.plot(df['50_MA'], label='50-Day MA')
plt.plot(df['200_MA'], label='200-Day MA')
plt.legend()
plt.show()
```

```python
df['Volume'].plot(title='Trading Volume')
plt.show()
```

```python
df['20_MA'] = df['Close'].rolling(window=20).mean()
df['BB_Upper'] = df['20_MA'] + (df['Close'].rolling(window=20).std() * 2)
df['BB_Lower'] = df['20_MA'] - (df['Close'].rolling(window=20).std() * 2)

plt.plot(df['Close'], label='Close Price')
plt.plot(df['BB_Upper'], label='Upper Band')
plt.plot(df['BB_Lower'], label='Lower Band')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df['Close'], model='additive', period=365)
decomposition.plot()
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
import matplotlib.pyplot as plt
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(df['Date'],inplace=True)

data = df['Close']
model = ARIMA(data,order=(5,1,0))
arima_results = model.fit()

print(arima_results.summary())

forecast = arima_results.forecast(steps=30)
print(forecast)

plt.plot(data, label='Historical Data')
plt.plot(forecast, label='Forecast', color='red')
plt.title('ARIMA Model Forecast')
plt.legend()
plt.show()

```
