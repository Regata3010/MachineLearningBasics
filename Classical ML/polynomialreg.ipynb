```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(42)  
X = np.random.rand(20, 1) * 10  
y = 2 * X**2 + 3 * X + 5 + np.random.randn(20, 1) * 10 
```

```python
X
```

```python
y
```

```python
plt.scatter(X, y, color='blue')
plt.title("Generated Data")
plt.show()
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
print(X_poly[:1])
```

```python
model = LinearRegression()
model.fit(X_poly, y)

# Print the coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

```python
poly1 = PolynomialFeatures(degree=100)
X_poly1 = poly1.fit_transform(X)
model1  = LinearRegression()
model1.fit(X_poly1,y)
```

```python
X_range = np.linspace(0, 10, 100).reshape(-1, 1)
X_range_poly = poly1.transform(X_range)
y_pred = model1.predict(X_range_poly)

# Plot the original data and the polynomial curve
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X_range, y_pred, color='red', label='Polynomial Fit (Degree 100)')
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```
