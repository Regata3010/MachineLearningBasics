```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DailyDelhiClimateTest.csv')
df
```

```python
df.drop(columns = ['humidity','wind_speed','meanpressure'],inplace = True)
```

```python
df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
X1 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
df['ewma'] = X1
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],X1,color='black',label = 'alpha = 0.9')
plt.legend()
plt.show()
```

```python
X2 = df['meantemp'].ewm(alpha=0.1).mean()
```

```python
df['ewma0.1'] = X2
df
```

```python
plt.scatter(df['date'],df['meantemp'],color = 'green')
plt.plot(df['meantemp'],X2,label = 'alpha = 0.1')
plt.legend()
plt.show()
```

```python
X3 = df['meantemp'].ewm(alpha=0.5).mean()
```

```python
df['ewma0.5'] = X3
df
```

```python
plt.scatter(df['date'],df['meantemp'],color = 'orange')
plt.plot(df['meantemp'],X3,label = 'alpha = 0.5')
plt.legend()
plt.show()
```
