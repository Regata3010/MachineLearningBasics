```python
items = ["A", "B", "C", "D"]
weights = [0.2, 0.3, 0.1, 0.4] #biased sampling means the more the weight more the probability of occuring the resultant list
sample_size =3
```

```python
sample_items = random.choices(items,weights=weights,k=sample_size)
print(sample_items)
```

```python
individuals = [("A", 0.2), ("B", 0.3), ("C", 0.4), ("D", 0.1)]
cumulative_fitness = [sum([ind[1] for ind in individuals[:i+1]]) for i in range(len(individuals))]
cum_fit = [sum(l[1] for l in individuals[:i+1]) for i in range(len(individuals))]
print(cumulative_fitness)
print(cum_fit)
```
